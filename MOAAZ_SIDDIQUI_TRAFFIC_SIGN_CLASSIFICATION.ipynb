```python
test_datagen = ImageDataGenerator(rescale=1/255)

test_generator = test_datagen.flow_from_directory(
        '/content/TRAFFIC DATA/TEST',  
        target_size=(144, 131), 
        batch_size=64,
        class_mode='categorical')
```

```python
class myCallback(tf.keras.callbacks.Callback):
  def on_epoch_end(self, epoch, logs={}):
    if(logs.get('acc')>0.97):
      print("\nReached 97% accuracy so cancelling training!")
      self.model.stop_training = True

callbacks = myCallback()
```

```python
model = tf.keras.models.Sequential([tf.keras.layers.Conv2D(16, (3,3), input_shape=(144,131,3), activation = 'relu'),
                                   tf.keras.layers.MaxPooling2D(2,2), 
                                    tf.keras.layers.Conv2D(32, (3,3), activation = 'relu'),
                                    tf.keras.layers.MaxPooling2D(2,2),
                                    tf.keras.layers.Dropout(0.25),
                                    tf.keras.layers.Conv2D(32, (3,3), activation = 'relu'),
                                    tf.keras.layers.MaxPooling2D(2,2),
                                    tf.keras.layers.Dropout(0.25),
                                    tf.keras.layers.Flatten(), 
                                    tf.keras.layers.Dense(256, activation = 'relu'),
                                    tf.keras.layers.Dropout(0.25),
                                    tf.keras.layers.Dense(256, activation = 'relu'),
                                
                                    tf.keras.layers.Dense(58, activation = 'softmax')])
```

```python
model.compile(loss='categorical_crossentropy', optimizer=Adam(lr = 0.003), metrics=['acc'])
```

```python
history = model.fit(
      train_generator,
      steps_per_epoch=10,  
      epochs=25, callbacks=callbacks) 
```

```python
score = model.evaluate(test_generator, batch_size=32, verbose = 2)
```
